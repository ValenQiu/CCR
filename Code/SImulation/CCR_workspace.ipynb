```python
import numpy as np
import matplotlib.pyplot as plt

arc_length = 30  # Length of the arc, cm
angles = np.linspace(-np.pi/2, np.pi/2, 30)
curvatures = angles / arc_length

# Create the figure and axis with a fixed size
fig, ax = plt.subplots(figsize=(5, 5))  # Set the figure size

for i, curvature in enumerate(curvatures):
    # Calculate the start and end angles of the arc
    if curvature == 0:
        # If the curvature is 0, the arc is a straight line
        y = np.linspace(0, arc_length, 100)
        x = np.zeros_like(y)
        end_x, end_y = 0, arc_length  # End point of the straight line
    else:
        # Otherwise, create the arc coordinates
        theta = np.linspace(0, arc_length * curvature, 100)
        x = np.cos(theta) / curvature
        y = np.sin(theta) / curvature
        
        # Shift the arc to start at (0, 0)
        x = x - x[0]
        y = y - y[0]
        
        # Calculate the end point of the arc
        end_x = x[-1]
        end_y = y[-1]

    # Set the alpha value based on the arc's position
    alpha_value = 0.8 if (i == 0 or i == len(curvatures) - 1) else 0.5

    # Plot the arc with the specified alpha
    ax.plot(x, y, color='blue', linewidth=2, alpha=alpha_value)

    # Calculate the tangent angle at the end of the arc using the modified formula
    if curvature != 0:
        tangent_angle = np.pi / 2 + curvature * arc_length  # Calculate the tangent angle
        
    else:
        tangent_angle = np.pi / 2  # For straight line, vertical tangent

    # Length of the tangent line
    tangent_length = arc_length / 5
    
    print("tangent angle: ",tangent_angle)

    # Calculate the tangent line end points
    tangent_start_x = end_x
    tangent_start_y = end_y
    tangent_end_x = tangent_start_x + tangent_length * np.cos(tangent_angle)
    tangent_end_y = tangent_start_y + tangent_length * np.sin(tangent_angle)

    # Plot the tangent line as a dotted line
    ax.plot([tangent_start_x, tangent_end_x], 
            [tangent_start_y, tangent_end_y], 
            color='blue', linestyle='dotted', alpha=0.3, linewidth=2)

# Set the aspect ratio to 'equal' to ensure the arcs appear circular
ax.set_aspect('equal')

# Set the x and y limits
ax.set_xlim(-25, 25)
ax.set_ylim(0, 50)

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Arcs with Tangent Dotted Lines')

# Show the plot
plt.show()
```
