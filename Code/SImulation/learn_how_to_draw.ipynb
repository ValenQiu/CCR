```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 定义圆的参数
radius = 12.5
theta = np.linspace(0, 2 * np.pi, 100)
r = np.linspace(0, radius, 100)
theta, r = np.meshgrid(theta, r)

# 计算圆的坐标
x = r * np.cos(theta)
y = r * np.sin(theta)
z = np.zeros_like(x)  # z 坐标为 0，表示在 xy 平面上

# 创建画布
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制圆形平面
ax.plot_surface(x, y, z, color='b', alpha=0.6)

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 设置坐标轴比例相同
ax.set_box_aspect([1,1,1])  # 设置坐标轴比例

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 定义圆的参数
radius = 12.5
theta = np.linspace(0, 2 * np.pi, 20)
r = np.linspace(0, radius, 20)
theta, r = np.meshgrid(theta, r)

# 计算圆的坐标
x = r * np.cos(theta)
y = r * np.sin(theta)

# 创建画布
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制五个圆形平面，每个平面之间的 z 轴距离为 3
for i in range(5):
    z = np.full_like(x, i * 30)  # 设置 z 坐标
    ax.plot_surface(x, y, z, color='b', alpha=0.6)

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 设置坐标轴比例相同
ax.set_box_aspect([1,1,6])  # 设置坐标轴比例

# 显示图形
plt.show()

```
